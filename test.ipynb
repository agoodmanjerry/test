```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
intervals = np.resize(np.sort(11 * np.random.rand(2*10)), (10,2))
counts = np.sort(12*np.random.rand(7))

print(intervals)
print(counts)
```

```python
within = []
for i in range(len(intervals)):
    a = (intervals[i][0] < counts)
    b = (intervals[i][1] > counts)
    # print(a&b)
    within.extend(counts[a & b].tolist())

print(within)
```

```python
def x(t, a):
    return 20*np.sin(2*np.pi*1000*t - np.pi/3)*np.exp(-a*t)

a_list = [500, 750, 1000]
step = 1e-7
t = np.arange(-2e-3, 2e-3 + step, step=step)

plt.figure(figsize=(20,6))
for a in a_list:
    plt.plot(t, x(t,a), label=f'a = {a}')

plt.title('Damping Sin')
plt.xlabel('t (s)')
plt.ylabel('x(t)')
plt.legend()
# plt.savefig('')
plt.show()
```
